```python
filters=Gowdown[['Muncipal_Corporation', 'Food_gowdown', 'Anna_canteen','Food_temperature', 'Total_quantity','Route_steps','Delivery_location','Delivery_time','Employee_details']]
filters
```

```python
df=pd.DataFrame(Gowdown)
```

```python
def Food_checking_app(df,Muncipal_Corporation,Food_gowdown,Anna_canteen):
    newdf=filters[(filters["Muncipal_Corporation"]==Muncipal_Corporation) & (filters["Food_gowdown"]==Food_gowdown) & (filters["Anna_canteen"]==Anna_canteen)]
    final=newdf.to_dict('dict')
    return final
```
